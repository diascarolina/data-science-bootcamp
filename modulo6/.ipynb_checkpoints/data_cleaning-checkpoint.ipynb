```python
import pandas as pd
```

```python
url_clientes_aprovados = 'https://github.com/diascarolina/data-science-bootcamp/blob/main/data/clientes_aprovados.csv?raw=true'

url_clientes_cadastrados = 'https://github.com/diascarolina/data-science-bootcamp/blob/main/data/clientes_cadastrados.csv?raw=true'
```

```python
clientes_aprovados = pd.read_csv(url_clientes_aprovados)

clientes_cadastrados = pd.read_csv(url_clientes_cadastrados)
```

```python
clientes_aprovados.head()
```

```python
aproved_clients = clientes_aprovados.copy()
aproved_clients.columns = ['id', 'month', 'status']
```

```python
aproved_clients['status'].unique()
```

```python
status_dict = {
    'nenhum empréstimo': 'no_loan',
    '1-29 dias': '1-29days',
    'pagamento realizado': 'paid_loan',
    '30-59 dias': '30-59days',
    '60-89 dias': '60-89days',
    '90-119 dias': '90-119days',
    '120-149 dias': '120-149days',
    '>150 dias': '>150days'
}

aproved_clients['status'] = aproved_clients['status'].map(status_dict)
```

```python
aproved_clients['status'].unique()
```

```python
aproved_clients
```

```python
client_record = clientes_cadastrados.copy()
```

```python
client_record.head()
```

```python
client_record.columns
```

```python
client_record.columns = ['id', 'gender', 'own_car', 'own_property', 'children_count', 'annual_income',
                        'income_type', 'education_type', 'marital_status', 'housing_type',
                        'age', 'own_cellphone', 'own_workphone', 'own_phone', 'own_email',
                        'occupation_type', 'family_size', 'years_working']
```

```python
client_record
```

```python
gender_dict = {
    'H': 'm',
    'M': 'f',
    }

client_record['gender'] = client_record['gender'].map(gender_dict)
```

```python
client_record['income_type'].unique()
```

```python
income_type_dict = {
    'Empregado': 'working',
    'Associado comercial': 'commercial_associate',
    'Pensionista': 'pensioner',
    'Servidor publico': 'state_servant',
    'Estudante': 'student'
    }

client_record['income_type'] = client_record['income_type'].map(income_type_dict)
```

```python
client_record['education_type'].unique()
```

```python
education_type_dict = {
    'Ensino superior': 'higher_education',
    'Ensino medio': 'secondary',
    'Ensino superior incompleto': 'incomplete_higher',
    'Ensino fundamental': 'lower_secondary',
    'Pos graduacao': 'academic_degree'
}

client_record['education_type'] = client_record['education_type'].map(education_type_dict)
```

```python
client_record['marital_status'].unique()
```

```python
marital_status_dict = {
    'Uniao estavel': 'civil_marriage',
    'Casado': 'married',
    'Solteiro': 'single',
    'Separado': 'separated',
    'Viuvo': 'widow'
}

client_record['marital_status'] = client_record['marital_status'].map(marital_status_dict)
```

```python
client_record['housing_type'].unique()
```

```python
housing_type_dict = {
    'Alugado': 'rented_apartment',
    'Casa propria': 'house_apartment',
    'Habitacao publica': 'municipal_apartment',
    'Mora com os pais': 'with_parents',
    'Cooperativa habitacional': 'coop_apartment',
    'Apartamento comercial': 'office_apartment'
}

client_record['housing_type'] = client_record['housing_type'].map(housing_type_dict)
```

```python
client_record['occupation_type'].unique()
```

```python
occupation_type_dict = {
    'Outros': 'others',
    'Seguranca': 'security',
    'Venda': 'sales',
    'Contabilidade': 'accounting',
    'Construcao civil': 'civil_construction',
    'Gerencia': 'management',
    'Motorista': 'driver',
    'Equipe principal': 'principal_team',
    'Alta tecnologia': 'high_tech',
    'Limpeza': 'cleaning',
    'Servico privado': 'private_sector',
    'Cozinha': 'kitchen',
    'Baixa qualificacao': 'low_skilled',
    'Medicina': 'medicine',
    'Secretariado': 'secretary',
    'Garcom': 'waiter',
    'RH': 'hr',
    'Corretor imobiliario': 'realtor',
    'TI': 'it'
}

client_record['occupation_type'] = client_record['occupation_type'].map(occupation_type_dict)
```
